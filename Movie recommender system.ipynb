```python
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name']);
    return L
```

```python
#we have to do the same thing to the keywords as welll as the cast and crew data of the datasets
movies['keywords']
```

```python
def convertk(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['keywords']=movies['keywords'].apply(convertk)
```

```python
movies.head()
```

```python
#lets do the exact same for the cast and the crew portion of the dataset too
#but here we only need the 3 (i.e top 3 actors name so we have to set something as an actor thing, here how it goes brother
```

```python
def convert3(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
#to look for the crew part, what we have to do is, check if the job is director and then store the name of it
```

```python
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
ps.stem('actioned')
```

```python
#we're making a funciton so that we can use this porterstemmer
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))

    return " ".join(y)
```
