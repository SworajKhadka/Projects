```python
#alright, so the page thinks that we are bot or sth which causes the response to be denied, hence we have to show them that this is actually coming from a user agent, therefore we use an user-agent
```

```python
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 6.3; Win64; x64) "
                  "AppleWebKit/537.36 (KHTML, like Gecko) "
                  "Chrome/120.0.0.0 Safari/537.36",
    "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,*/*;q=0.8",
    "Accept-Language": "en-US,en;q=0.5",
    "Connection": "keep-alive"
}

webpage=requests.get('https://www.ambitionbox.com/companies-in-bengaluru',headers=headers).text
 
```
